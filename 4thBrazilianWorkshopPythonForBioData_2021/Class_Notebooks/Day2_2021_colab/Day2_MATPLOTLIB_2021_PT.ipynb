#**MatplotLib**

Matplotlib é uma biblioteca do Python utilizada para visualização de dados e criação de gráficos gráficos 2D a partir de dados em arrays.


###**Pyplot**

É um módulo da biblioteca Matplotlib que oferece uma interface semelhante ao MATLAB (programa interativo que se destina a cálculos numéricos e gráficos científicos).

###**Matplotlib.pyplot**
É uma coleção de funções que fazem o matplotlib funcionar como MATLAB. Cada função do pyplot cria ou altera algo na figura. Indo desde funções que criam a figura e a área de plotagem até funções que plotem linhas, rótulos (labels) etc.

###**Partes de uma figura**

A figura construída pelo Matplotlib é divida em algumas partes e entendê-las é importante para termos o domínio de construção e identificar funções atribuídas a cada uma delas. Essas partes são:


Figure: Uma figura inteira que pode conter um ou mais axes (plots).

Axes: Geralmente pensado como plot. Cada axes tem um title, x-lable e y-lable.

Axis: Responsável por determinar os limites do gráfico.

A seguir, criaremos alguns gráficos com Matplotlib utilizando dados do artigo do Dr. José Bruno Malaquias sobre dispersão larval de cepas (Resistente e Suscetível) de uma espécie de lagarta (Spodoptera frugiperda) em duas variedades de algodão (WS e NBT).


**Contextualização:**

A lagarta-do-cartucho, como é conhecida a Spodoptera frugiperda, é um inseto praga presente em sistemas de cultivo que possui uma grande capacidade de dispersão durante o período de cultivo e uma ampla gama de hospedeiros, incluindo o algodão.

Ao longo dos anos, vários pesquisadores tem investigado soluções para o combate desses insetos praga. Uma das alternativas que tem chamado atenção é o uso de bactéricas patogênicas com potencial controle de pragas agrícolas e urbanas, entre essas a bactéria Bacillus thuringiensis (Bt) tem se destacado.

A eficácia e especificidade das cepas de Bt e suas toxinas no controle de insetos praga, favoreceu a formulação de biopesticidas à base deste patógeno. Além disso, posteriormente, os genes que codificam as toxinas de ação inseticida dessas bactérias foram também isolados e incorporados em plantas, gerando plantas transgênicas Bt.

Nesse contexto, algodões Bt passaram a ser produzidos a fim de combater insetos praga, incluindo a lagarta S. frugiperda. Entretanto, nas últimas décadas tem ocorrido falha no controle dessa praga por esse sistema.

Uma das evidências da falha é a contaminação cruzada entre diferentes espécies de plantas ou fluxo gênico entre as variedades Bt e non-Bt, em que dependendo do grau de contaminação e se as larvas de S. frugiperda se movem entre as plantas de um campo, a geração de larvas resistentes pode ser acelerada.

Uma outra evidência para sobrevivência larval em plantas Bt é o breve tempo de exposição durante o qual o inseto não ingere uma quantidade suficiente de proteínas inseticidas para causar mortalidade antes da metaformose para o estado de pupa.

Além disso, plantas modificadas geneticamente ao serem resistentes a insetos podem afetar o comportamento alimentar, locomoção e dispersão das larvas desses insetos.

Pensando em todas essas questões, o objetivo do estudo foi:

1) Investigar o comportamento alimentar de cepas resistentes e suscetíveis de Spodoptera frugiperda (lagarta) nas variedades de algodões Bt e non-Bt.

2) Investigar o efeito do movimento larval na sua distribuição populacional.


**Referência:**

Malaquias, J. B., Godoy, W. A., Garcia, A. G., Ramalho, F. D. S., & Omoto, C. (2017). Larval dispersal of Spodoptera frugiperda strains on Bt cotton: a model for understanding resistance evolution and consequences for its management. Scientific reports, 7(1), 1-10.

O primeiro passo é determinar quantas amostras suscetíveis e resistentes há no grupo de taxa de dispersão acima ou igual a 0.5 e no grupo com taxa abaixo de 0.5.

Uma das formas de descorbrir esses valores será apresentada a seguir:

1)Filtre as amostras com taxa de dispersão acima ou igual a 0.5

`a = dataset[dataset['Dispersao'] >= 0.5]` -> Filtra os dados de dispersão acima ou igual a 0.5

2) Em seguida, conte quantos Res e Sus possuem na coluna Linhagem utilizando a função `groupby()` e `size()`

`a.groupby(['Linhagem']).size()` -> Filtra a coluna 'Linhagem'e conta número de Res e Sus

3) Converta os dados para uma lista com a função `.tolist()`

`a.groupby(['Linhagem']).size().tolist()` -> Valores convertidos para uma lista

````
#Filtre os dados de dispersão acima ou igual a 0.5
dispersao_acima_05 = dataset[dataset['Dispersao'] >= 0.5]
dispersao_acima_05.head()
````

````
#Conte quantas amostras são Sus e quantas são Res na coluna 'Linhagem'
dispersao_acima_05.groupby(['Linhagem']).size()
````

````
#Converta a contagem de Res e Sus para uma lista
list_res_sus_acima_05 = dispersao_acima_05.groupby(['Linhagem']).size().tolist()
list_res_sus_acima_05
````

Faça o mesmo processo anterior para o grupo com taxa de dispersão abaixo de 0.5:

1)Filtre as amostras com taxa de dispersão abaixo de 0.5

`a = dataset[dataset['Dispersao'] < 0.5]` -> Filtra os dados de dispersão abaixo de 0.5

2) Em seguida, conte quantos Res e Sus possuem na coluna
Linhagem utilizando a função `groupby()` e `size()`

`a.groupby(['Linhagem']).size()` -> Conta número de Res e Sus na coluna linhagem

3) Converta os dados para uma lista com a função `.tolist()`

`a.groupby(['Linhagem']).size().tolist()` -> Valores convertidos para uma lista

````
#Filtre os dados de dispersão abaixo de 0.5
dispersao_abaixo_05 = dataset[dataset['Dispersao'] < 0.5]
dispersao_abaixo_05.head()
````

````
#Conte quantas amostras são Sus e quantas são Res
dispersao_abaixo_05.groupby(['Linhagem']).size()
````

````
#Converta a contagem de Res e Sus para uma lista
list_res_sus_abaixo_05 = dispersao_abaixo_05.groupby(['Linhagem']).size().to_list()
list_res_sus_abaixo_05
````

A quantidade de linhagens resistentes (Res) e suscetíveis (Sus) no grupo com taxa de dispersão acima ou igual a 0.5 e no grupo com abaixo de 0.5 estão nas listas geradas separadamente pra cada grupo. Porém, podemos colocar esses resultados em uma única lista para facilitar seu uso posteriormente.

````
#Crie uma lista com o resultado das duas listas anteriormente.

list_res_sus = list_res_sus_acima_05 + list_res_sus_abaixo_05
list_res_sus
````

Com os dados de input já prontos. Podemos usar a função pie() para criar os gráficos de pizza.

Será criado um gráfico de pizza que represente o grupo com taxa de dispersão acima ou igual a 0.5 e o grupo abaixo de 0.5.
````
#Crie a figura e área de plot
fig, ax = plt.subplots(1,1)

#Crie o gráfico de pizza para taxa de dispersão (>=0.5 e <0.5)
plt.pie(x = quantidade_amostras, #Quantidade de amostras
        labels = ['Dispersão >= 0.5', 'Dispersão < 0.5'], #Nome dos grupos
        colors = ['darkslategray', 'maroon'], #Cores para cada grupo
        startangle = 90, #Define o ângulo que o gráfico inicia
        )

#Crie o gráfico de pizza para Linhagem (Res e Sus)
plt.pie (x = list_res_sus, #Quantidade de amostras
         colors = ['beige', 'azure'], #Cor de cada grupo
         radius = 0.65, #Determina o tamanho da pizza
         startangle = 90, #Determina o ângulo que o gráfico inicia
         labeldistance = 0.85, #Determina a distância do nome dos grupos em relação ao centro da pizza
         autopct = '%1.1f%%', #Apresenta os valores de cada grupo
         pctdistance = 0.5, #Determina a distância dos valores em relação ao centro da pizza

         )

#Crie a legenda
plt.legend(labels = ['>= 0.5', '< 0.5', 'Res', 'Sus'], #Nome dos grupos
           title = 'Legenda') #Título da legenda

#Identifique que os axis dos gráficos são iguais
plt.axis('equal')

#Defina um layout para o plot
plt.tight_layout ()

#Apresente a figura
plt.show()
````

#**Box plot**

O boz plot é um gráfico utilizado para visualizar a distribuição de dados de um variável numérica, incluindo outliers.

 Ele é formado por medidas descritivas, tais como mínimo, máximo, primeiro quartil, segundo quartil ou mediana e terceiro quartil.

Observação: Seus entalhes exibem o intervalo de confiança em torno da mediana.
Se os entalhes de duas caixas não se sobrepuserem, haverá uma 'forte evidência' -
com um nível de confiança de 95% - de que as medianas diferem.

#**Figura com a combinação de gráficos**

A seguir será criado uma única figura com o gráfico de barra, gráfico de pontos e gráfico de linhas da taxa de sobrevivência de acordo com os intervalos de tempo de exposição. Para isso utilizaremos a função `subplots()` para definirmos a localização dos vários plots em uma única figura.

Observação: Cada axes criado pela função subplots() corresponde a um plot. Logo, utilizaremos o comando axs[] para trabalhar com cada plot individualmente.

````
#Crie a figura e os plots
fig, axs = plt.subplots(nrows = 1, #indica o numéro de linhas
                        ncols = 3, #indica o número de colunas
                        figsize=(9, 3), #define o tamanho da figura
                        sharey=True) #indica que os plots compartilham o mesmo eixo y

#No axes zero(0) crie o gráfico de barra
axs[0].bar(x = tempo, #Grupos do eixo x
           height = medias, #Altura das barras
           width = 3 ) #Largura das barras
axs[0].set_xticks([6,12,18,24]) #Altera o nome das ticks do eixo x
axs[0].set_ylabel('Sobreviência') #Define a legenda do eixo x

#No axes um (1) crie o gráfico de pontos
axs[1].scatter(x = tempo, #Posição dos dados no eixo x
               y = medias) #Posição dos dados no eixo y
axs[1].set_xticks([6,12,18,24]) #Altera o nome das ticks do eixo x
axs[1].set_xlabel('Tempo (horas)') #Define a legenda do eixo x

#No axes dois (2) crie o gráfico de linhas
axs[2].plot(tempo, #Define posição da linha no eixo x
            medias, #Define posição da linha no eixo y
            marker='o') #Define o tipo de marcador
axs[2].set_xticks([6,12,18,24])#Define a legenda do eixo x

#Defina o título da figura
fig.suptitle('Sobrevivência ao longo do tempo')

#Apresente a figura
plt.plot ()
````

#**Salvar uma figura**

Para salvar figuras utilize a função `savefig()`. A partir dela podemos definir orientação, formato, resolução etc.

````
#Crie a figura
fig, ax = plt.subplots()

#Gráfico de barras
plt.bar(tempo,
        medias,
        width = 3,
        tick_label = [6,12,18,24])


#Legenda dos eixos
ax.set_xlabel('Tempo (horas)')
ax.set_ylabel('Taxa de sobrevivência')
fig.suptitle('Sobrevivência ao longo do tempo')

#Para salvar a figura
fig.savefig(fname = 'bar.jpg', #nome da figura
            dpi = 100, #resolução da figura
            format = 'jpg', #formato
            quality = 95, #qualidade
            facecolor = 'pink', #cor de fundo
            orientation = 'landscape') #orientação
````
