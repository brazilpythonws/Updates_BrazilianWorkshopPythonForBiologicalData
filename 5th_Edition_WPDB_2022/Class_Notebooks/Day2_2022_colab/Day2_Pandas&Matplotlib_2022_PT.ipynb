```python
df.columns = ['tipo',
              "%",
              "leituras",
              'taxon_id',
              'dominio',
              'filo',
              'classe',
              'ordem',
              'familia',
              'genero',
              'especie']
```

```python
print(df.groupby(by=['dominio'])['leituras'].sum())
```

```python
dominio = df.groupby(by=['dominio'])['leituras'].sum()
```

```python
dominio
```

```python
fig, ax = plt.subplots(figsize = (5,5))

plt.pie(x=dominio,
        labels = dominio.index,
        radius=1,
        center=(1, 1),
       wedgeprops={"linewidth": 1, "edgecolor": "red"},
        frame=False)

plt.show()
```

```python
fig, ax = plt.subplots(figsize = (5,5))
plt.pie(x=dominio, labels = dominio.index,
        radius = 1, center=(1,1),
        wedgeprops={"linewidth":1, "edgecolor": 'red'},
                    frame=False) #index indica que eu quero os valores
plt.show()
```

```python
fig, ax = plt.subplots()
plt.pie(x=dominio,
        labels= dominio.index,
        autopct = '%1.1f%%',
        radius=1.5,
        wedgeprops = {'linewidth':2},
        pctdistance = 0.8,
        labeldistance = 1.5,
        textprops={'fontsize':0}
        )

plt.legend(labels = dominio.index,
           title = "Dominio",
           fontsize = 16,
           title_fontsize = 18,
           bbox_to_anchor = (1,0,0.5,1))

plt.show()
```

```python
aceto = df.loc[(df['familia']== 'Acetobacteraceae')]
aceto_soma = aceto.groupby(by=['genero'])['leituras'].sum().reset_index()
aceto_soma
```

```python
fig, ax = plt.subplots()

ax.bar(aceto_soma['genero'],
       aceto_soma['leituras'],
       width = 1,
       edgecolor = "blue",
       linewidth = 0.7)

plt.xticks(rotation = 45, ha = 'right')

plt.show()
```

```python
aceto_soma['porcentagem'] = aceto_soma['leituras']/(aceto_soma['leituras'].sum()) * 100
```

```python
aceto_soma.head()
```

```python
aceto_soma_1 = aceto_soma[aceto_soma['porcentagem'] >= 1]
aceto_soma_1.head()
```

```python
fig, ax = plt.subplots()

ax.bar(aceto_soma_1['genero'],
       aceto_soma_1['leituras'],
       width = 1,
       edgecolor = 'red',
       linewidth = 0.7)

plt.xticks(rotation= 45, ha= 'right')

plt.show()
```

```python
arqueias = df.loc[(df['dominio']=="Archaea")]
arqueias
```

```python
arqueias_f = arqueias.groupby(by=['filo','ordem'])['leituras'].sum().unstack('ordem').reset_index()
arqueias_f
```

```python
arqueias.groupby(by=['filo','ordem'])['leituras'].sum().unstack('ordem').reset_index().plot(kind='bar', stack='True')
```

```python
fig = plt.figure(figsize=(5,5))

ax1 = arqueias_f.plot(kind='bar', width=0.8, stacked='True')
ax1.legend(loc = 'upper left', bbox_to_anchor=(1.05,1), fancybox = True, shadow = True, ncol=1)
plt.xlabel('filo')
plt.ylabel('Contagem total')
plt.xticks((0,1,2,3),('Crenarchaeota','Euryarchaeota','Nanoarchaeota', 'Thaumarchaeota'),rotation = 45)
```

```python
arqueias_f_t = arqueias_f.transpose()
arqueias_f_t
```

```python
arqueias_f_t.drop(arqueias_f_t.index[0], inplace = True)
```

```python
arqueias_f_t
```

```python
arqueias_f_t_p = arqueias_f_t/arqueias_f_t[arqueias_f_t.columns].sum() *100

arqueias_f_t_p
```

```python
fig = plt.figure(figsize=(5,5))
ax1 = arqueias_f_t_p.transpose().plot(kind ='bar',width = 0.8, stacked= "True")
ax1.legend(loc="upper left", bbox_to_anchor=(1.05,1),fancybox = True, shadow = True, ncol=1)
plt.xlabel('Filo')
plt.ylabel('% de leituras')
plt.xticks((0,1,2,3),('Crenarchaeota','Euryarchaeota','Nanoarchaeota', 'Thaumarchaeota'),rotation = 45)
```
