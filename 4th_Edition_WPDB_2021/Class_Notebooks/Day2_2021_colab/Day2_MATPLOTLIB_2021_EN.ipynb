#**Matplotlib**

Matplotlib is a Python library used for data visualization and creating 2D graphics from data in arrays.

###**Pyplot**

It is a module of the Matplotlib library that provides an interface similar to MATLAB (an interactive program designed for numerical calculations and scientific graphics).

###**Matplotlib.pyplot**

It is a collection of functions that make Matplotlib work like MATLAB. Each pyplot function creates or modifies something in the figure, ranging from functions that create the figure and plot area to functions that plot lines, labels, etc.

###**Parts of a Figure**

The figure created by Matplotlib is divided into several parts, and understanding them is important to have a grasp of its construction and to identify functions associated with each of them. These parts are:

- **Figure:** The entire figure that can contain one or more axes (plots).

- **Axes:** Generally thought of as a plot. Each axis has a title, x-label, and y-label.

- **Axis:** Responsible for determining the graph's limits.

Next, we will create some graphs with Matplotlib using data from Dr. Jos√© Bruno Malaquias' article on larval dispersion of strains (Resistant and Susceptible) of a caterpillar species (Spodoptera frugiperda) in two varieties of cotton (WS and NBT).

**Context:**

The fall armyworm, commonly known as Spodoptera frugiperda, is a pest insect in farming systems with a high dispersal capacity during the cultivation period and a wide range of hosts, including cotton.

Over the years, several researchers have investigated solutions to combat these pest insects. One alternative that has gained attention is the use of pathogenic bacteria with potential for controlling agricultural and urban pests, including the bacterium Bacillus thuringiensis (Bt).

The effectiveness and specificity of Bt strains and their toxins in controlling pest insects have led to the formulation of biopesticides based on this pathogen. Furthermore, the genes encoding the insecticidal action toxins of these bacteria have also been isolated and incorporated into plants, generating Bt transgenic plants.

In this context, Bt cottons were produced to combat pest insects, including the fall armyworm S. frugiperda. However, in recent decades, there have been failures in controlling this pest through this system.

One piece of evidence of the failure is the cross-contamination between different plant species or gene flow between Bt and non-Bt varieties. Depending on the degree of contamination and if the S. frugiperda larvae move between plants in a field, the generation of resistant larvae can be accelerated.

Another piece of evidence for larval survival on Bt plants is the brief exposure time during which the insect does not ingest a sufficient amount of insecticidal proteins to cause mortality before metamorphosing into the pupal stage.

Additionally, genetically modified plants that are resistant to insects can affect the feeding behavior, movement, and dispersion of these insect larvae.

Considering all these issues, the study aimed to:

1) Investigate the feeding behavior of resistant and susceptible strains of Spodoptera frugiperda (caterpillar) in Bt and non-Bt cotton varieties.

2) Investigate the effect of larval movement on their population distribution.

**Reference:**

Malaquias, J. B., Godoy, W. A., Garcia, A. G., Ramalho, F. D. S., & Omoto, C. (2017). Larval dispersal of Spodoptera frugiperda strains on Bt cotton: a model for understanding resistance evolution and consequences for its management. Scientific reports, 7(1), 1-10.

# **Pie Chart**

The pie chart, also known as a sector chart, is used for nominal qualitative data, where the slices represent data categories and their sizes correspond to their frequencies.

**Application**

Next, we will use the pie chart to represent the number of samples that had a dispersion rate above or equal to 0.5 and samples that had a rate below 0.5.


**Construction**

The first step is to find out how many samples have a dispersion rate above 0.5 and how many have a rate below 0.5.

````
# Number of samples with dispersion rate above or equal to 0.5

'''
dataset['dispersion'] >= 0.5  -> Returns a dataframe with True for dispersion rate above or equal to 0.5 and False for below 0.5
dataset[dataset['dispersion'] >= 0.5]   -> Returns a complete dataframe only with samples above or equal to 0.5
len()   -> Counts the number of rows. In this case, the number of rows equals the number of samples with rate above or equal to 0.5
'''

dispersion_count_above_05 = len(dataset[dataset['dispersion'] >= 0.5])
dispersion_count_above_05
````

````
# Number of samples with dispersion rate below 0.5

'''
dataset['dispersion'] > 0.5  -> Returns a dataframe with True for dispersion rate below 0.5 and False for above 0.5
dataset[dataset['dispersion'] > 0.5]   -> Returns a complete dataframe only with samples below 0.5
len()   -> Counts the number of rows. In this case, the number of rows equals the number of samples with rate below 0.5
'''

dispersion_count_below_05 = len(dataset[dataset['dispersion'] < 0.5])
dispersion_count_below_05
````

````
# Put in a list the number of samples from the group with dispersion rate above or equal to 0.5 and below 0.5
sample_count = [dispersion_count_above_05, dispersion_count_below_05]
sample_count
````

Next, define the names of the groups that will be represented in the graph


````
# Determine the groups to be considered and put them in a list
group_names = ['dispersion >= 0.5', 'dispersion < 0.5']
group_names
````

Now use the `pie()` function to create a pie chart representing the number of samples in each dispersion rate group (above or equal to 0.5 and below 0.5).

 It's also possible to create two overlapping pie charts, each representing different classifications of the same data.

 In the following chart, we will represent the number of susceptible and resistant samples in groups with dispersion rate >= 0.5 and < 0.5.


**Construction**

The first step is to determine how many susceptible and resistant samples are there in the group with dispersion rate above or equal to 0.5 and in the group with rate below 0.5.

One way to find these values is presented below:

1) Filter samples with dispersion rate above or equal to 0.5

`a = dataset[dataset['Dispersao'] >= 0.5]` -> Filters data with dispersion above or equal to 0.5

2) Then, count how many Res and Sus are there in the 'Lineage' column using the `groupby()` function and `size()`

`a.groupby(['Lineage']).size()` -> Filters the 'Lineage' column and counts the number of Res and Sus

3) Convert the data to a list using the `.tolist()` function

`a.groupby(['Lineage']).size().tolist()` -> Values converted to a list






````
# Filter data with dispersion above or equal to 0.5
dispersion_above_05 = dataset[dataset['dispersion'] >= 0.5]
dispersion_above_05.head()
````

````
# Count how many samples are Sus and how many are Res in the 'Lineage' column
dispersion_above_05.groupby(['lineage']).size()
````

````
# Convert the count of Res and Sus to a list
list_res_sus_above_05 = dispersion_above_05.groupby(['lineage']).size().tolist()
list_res_sus_above_05
````

Do the same previous process for the group with dispersion rate below 0.5:

1) Filter samples with dispersion rate below 0.5

`a = dataset[dataset['Dispersion'] < 0.5]` -> Filters data with dispersion below 0.5

2) Then, count how many Res and Sus are there in the 'Lineage' column using the `groupby()` function and `size()`

`a.groupby(['Lineage']).size()` -> Counts the number of Res and Sus in the 'Lineage' column

3) Convert the data to a list using the `.tolist()` function

`a.groupby(['Lineage']).size().tolist()` -> Values converted to a list


````
# Filter data with dispersion below 0.5
dispersion_below_05 = dataset[dataset['dispersion'] < 0.5]
dispersion_below_05.head()
````

````
# Count how many samples are Sus and how many are Res
dispersion_below_05.groupby(['lineage']).size()
````

````
# Convert the count of Res and Sus to a list
list_res_sus_below_05 = dispersion_below_05.groupby(['lineage']).size().to_list()
list_res_sus_below_05
````

The quantity of resistant (Res) and susceptible (Sus) lineages in the group with dispersion rate above or equal to 0.5 and in the group with rate below 0.5 are stored in separate lists generated for each group. However, these results can be combined into a single list to facilitate their later use.


````
# Create a list with the result of the two lists generated earlier.

list_res_sus = list_res_sus_above_05 + list_res_sus_below_05
list_res_sus
````

With the input data ready, we can use the `pie()` function to create the pie charts.

One pie chart will represent the group with dispersion rate above or equal to 0.5, and the other will represent the number of resistant and susceptible samples within each dispersion rate group (>=0.5 and <0.5). Although these charts are created separately, they will be displayed together in a single axis.


````
# Create the figure and plot area
fig, ax = plt.subplots(1, 1)

# Create the pie chart for dispersion rate (>=0.5 and <0.5)
plt.pie(x=sample_count,  # Sample counts
        labels=['Dispersion >= 0.5', 'Dispersion < 0.5'],  # Group names
        colors=['darkslategray', 'maroon'],  # Colors for each group
        startangle=90  # Set the angle at which the pie chart starts
        )

# Create the pie chart for Lineage (Res and Sus)
plt.pie(x=list_res_sus,  # Sample counts
        colors=['beige', 'azure'],  # Color for each group
        radius=0.65,  # Determines the size of the pie chart
        startangle=90,  # Determines the angle at which the pie chart starts
        labeldistance=0.85,  # Determines the distance of group names from the center of the pie chart
        autopct='%1.1f%%',  # Displays the values for each group
        pctdistance=0.5,  # Determines the distance of values from the center of the pie chart
        )

# Create the legend
plt.legend(labels=['>= 0.5', '< 0.5', 'Res', 'Sus'],  # Group names
           title='Legend')  # Legend title

# Ensure the axes of the charts are equal
plt.axis('equal')

# Set a layout for the plot
plt.tight_layout()

# Display the figure
plt.show()
````

#**Box Plot**

The box plot is a graph used to visualize the distribution of data for a numeric variable, including outliers.

It consists of descriptive measures such as minimum, maximum, first quartile, second quartile (median), and third quartile.

Note: Its notches display the confidence interval around the median.
If the notches of two boxes do not overlap, there is 'strong evidence' -
with a 95% confidence level - that the medians differ.

## **Boxplot of Survival Rate by Cultivar**
**Question**

Is the survival rate of the WS cultivar different from the NBT cultivar?

**Application**

Boxplot depicting the survival rate according to cotton cultivars (WS and NBT).


**Construction**

The first step is to separate dataframes for each cotton cultivar:


````
# Dataframe only for the WS cultivar
cultivar_WS = dataset[dataset['cultivar'] == 'WS']
cultivar_WS.head()
````

````
dataset.columns
````

Instead of creating two separate boxplots for each group, we can place these boxplots in a single figure.

**Construction**

All parameters of the boxplot can be modified. Let's see some of these modifications below.


````
# Boxplot of the survival rate for WS and NBT cultivars in a single figure

# Create a figure and a plot area
fig = plt.figure()
ax = plt.subplot(1, 1, 1)

# Create a boxplot
plt.boxplot(survival_ws_nbt,
            labels=('WS', 'NBT'),
            vert=True,  # Vertical boxplot
            notch=True,  # Apply notching to the boxes
            positions=(1, 1.5),  # Positions of the boxplots
            widths=0.3,  # Width of the boxplots
            patch_artist=True,  # Fill with color
            showcaps=True,  # Show horizontal lines at the bottom and top of the box
            showbox=True,  # Show the box of the boxplot
            showfliers=True,  # Show outliers
            showmeans=True,  # Show means
            capprops=dict(color='orange'),  # Change color of lines connecting the base of the box to the lower limit and the top to the upper limit
            whiskerprops=dict(color='orange'),  # Change color of the horizontal lines representing the lower and upper limits
            flierprops=dict(markeredgecolor='black', marker='o', markersize=12, markerfacecolor='blue'),  # Outliers customization
            medianprops=dict(linewidth=2, color='red', linestyle='--'),  # Median line customization
            meanprops=dict(marker='D', markeredgecolor='red', markerfacecolor='red'),  # Mean line/marker customization
            boxprops=dict(facecolor='darkblue')  # Color the box of the boxplot
            )

# Modify ticks
# Modify ticks on the x-axis
plt.xticks(ticks=(1, 1.5),  # Set the location of the x-axis ticks
           labels=['Bt', 'non-Bt'],  # Set the x-axis labels
           fontsize=15)  # Set the label size

# Modify ticks on the y-axis
plt.yticks(fontsize=15)  # Set the label size

# Remove the top and right spines of the boxplots
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

# Add y-axis label
plt.ylabel('Survival Rate', fontdict={'size': 15})

plt.show()
````

In the previous plot, we changed the color of the boxplots. However, the chosen color was applied to both boxplots.
Next, we will see how to individually modify the colors of each boxplot.

````
# Same as the previous plot, but modifying the color of the boxplots individually

# Create the figure and a plot area
fig = plt.figure()
ax = plt.subplot(1,1,1)

bplot = plt.boxplot(survival_ws_nbt, labels=('WS', "NBT"),
                    vert=True,  # Vertical boxplot
                    notch=True,  # Apply notching to the boxes
                    positions=(1, 1.5),  # Positions of the boxplots
                    widths=0.3,  # Width of the boxplots
                    patch_artist=True,  # Fill with color
                    showcaps=True,  # Show horizontal lines at the bottom and top of the box
                    showbox=True,  # Show the box of the boxplot
                    showfliers=True,  # Show outliers
                    showmeans=True,  # Show means
                    capprops=dict(color='orange'),  # Change color of lines connecting the base of the box to the lower limit and the top to the upper limit
                    whiskerprops=dict(color='orange'),  # Change color of the horizontal lines representing the lower and upper limits
                    flierprops=dict(markeredgecolor='black', marker='o', markersize=12, markerfacecolor='blue'),  # Outliers customization
                    medianprops=dict(linewidth=2, color='red', linestyle='--'),  # Median line customization
                    meanprops=dict(marker='D', markeredgecolor='red', markerfacecolor='red')  # Mean line/marker customization
            )

# Set different colors for each boxplot
colors = ['pink', 'aquamarine']
for patch, color in zip(bplot['boxes'], colors):
    patch.set_facecolor(color)

# Modify ticks
plt.xticks(ticks=(1, 1.5), labels=['Bt', 'non-Bt'], fontsize=15)
plt.yticks(fontsize=15)

# Display the figure
plt.show()
````

# **Figure with Combination of Plots**

Next, a single figure will be created with a bar chart, scatter plot, and line chart of the survival rate according to the exposure time intervals. To achieve this, we will use the `subplots()` function to define the location of the various plots in a single figure.

Note: Each axes created by the `subplots()` function corresponds to a plot. Therefore, we will use the command axs[] to work with each plot individually.

````
# Create the figure and plots
fig, axs = plt.subplots(nrows=1, # Indicates the number of rows
                        ncols=3, # Indicates the number of columns
                        figsize=(9, 3), # Defines the size of the figure
                        sharey=True) # Indicates that the plots share the same y-axis

# In axes zero(0) create the bar chart
axs[0].bar(x=time, # Groups on the x-axis
           height=means, # Height of the bars
           width=3) # Width of the bars
axs[0].set_xticks([6,12,18,24]) # Change the names of the x-axis ticks
axs[0].set_ylabel('Survival') # Set the label for the x-axis

# In axes one (1) create the scatter plot
axs[1].scatter(x=time, # Position of the data on the x-axis
               y=means) # Position of the data on the y-axis
axs[1].set_xticks([6,12,18,24]) # Change the names of the x-axis ticks
axs[1].set_xlabel('Time (hours)') # Set the label for the x-axis

# In axes two (2) create the line chart
axs[2].plot(time, # Define the position of the line on the x-axis
            means, # Define the position of the line on the y-axis
            marker='o') # Define the marker type
axs[2].set_xticks([6,12,18,24]) # Set the label for the x-axis

# Set the title of the figure
fig.suptitle('Survival over Time')

# Show the figure
plt.plot()
````
