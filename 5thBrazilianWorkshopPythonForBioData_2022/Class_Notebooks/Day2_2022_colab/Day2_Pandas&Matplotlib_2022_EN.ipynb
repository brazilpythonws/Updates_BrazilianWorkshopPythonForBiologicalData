```python
df = pd.read_excel('/content/Barajas-2020/Tables/kaiju-metagenome-assignments.xlsx')
```

```python
df.head(1)
```

```python
df.columns = ['type',
              "%",
              "reads",
              'taxon_id',
              'domain',
              'phylum',
              'class',
              'order',
              'family',
              'genus',
              'species']
```

```python
print(df.groupby(by=['domain'])['reads'].sum())
```

```python
domain = df.groupby(by=['domain'])['reads'].sum()
```

```python
domain
```

```python
fig, ax = plt.subplots(figsize = (5,5))

plt.pie(x=domain,
        labels = domain.index,
        radius=1,
        center=(1, 1),
       wedgeprops={"linewidth": 1, "edgecolor": "red"},
        frame=False)

plt.show()
```

```python
fig, ax = plt.subplots(figsize = (5,5))
plt.pie(x=domain, labels = domain.index,
        radius = 1, center=(1,1),
        wedgeprops={"linewidth":1, "edgecolor": 'red'},
                    frame=False)
```

```python
fig, ax = plt.subplots()
plt.pie(x=domain,
        labels= domain.index,
        autopct = '%1.1f%%',
        radius=1.5,
        wedgeprops = {'linewidth':2},
        pctdistance = 0.8,
        labeldistance = 1.5,
        textprops={'fontsize':0}
        )

plt.legend(labels = domain.index,
           title = "Domain",
           fontsize = 16,
           title_fontsize = 18,
           bbox_to_anchor = (1,0,0.5,1))

plt.show()
```

```python
aceto = df.loc[(df['family']== 'Acetobacteraceae')]
aceto_sum = aceto.groupby(by=['genus'])['reads'].sum().reset_index()
aceto_sum
```

```python
fig, ax = plt.subplots()

ax.bar(aceto_sum['genus'],
       aceto_sum['reads'],
       width = 1,
       edgecolor = "blue",
       linewidth = 0.7)

plt.xticks(rotation = 45, ha = 'right')

plt.show()
```

```python
aceto_sum['porcent'] = aceto_sum['reads']/(aceto_sum['reads'].sum()) * 100
```

```python
aceto_sum.head()
```

```python
aceto_sum_1 = aceto_sum[aceto_sum['porcent'] >= 1]
aceto_sum_1.head()
```

```python
fig, ax = plt.subplots()

ax.bar(aceto_sum_1['genus'],
       aceto_sum_1['reads'],
       width = 1,
       edgecolor = 'red',
       linewidth = 0.7)

plt.xticks(rotation= 45, ha= 'right')

plt.show()
```

```python
arqueias = df.loc[(df['domain']=="Archaea")]
arqueias
```

```python
arqueias_f = arqueias.groupby(by=['phylum','order'])['reads'].sum().unstack('order').reset_index()
arqueias_f
```

```python
fig = plt.figure(figsize=(5,5))

ax1 = arqueias_f.plot(kind='bar', width=0.8, stacked='True')
ax1.legend(loc = 'upper left', bbox_to_anchor=(1.05,1), fancybox = True, shadow = True, ncol=1)
plt.xlabel('phylum')
plt.ylabel('Total count')
plt.xticks((0,1,2,3),('Crenarchaeota','Euryarchaeota','Nanoarchaeota', 'Thaumarchaeota'),rotation = 45)
```

```python
arqueias_f_t = arqueias_f.transpose()
arqueias_f_t
```

```python
arqueias_f_t.drop(arqueias_f_t.index[0], inplace = True)
```

```python
arqueias_f_t
```

```python
arqueias_f_t_p = arqueias_f_t/arqueias_f_t[arqueias_f_t.columns].sum() *100

arqueias_f_t_p
```

```python
fig = plt.figure(figsize=(5,5))
ax1 = arqueias_f_t_p.transpose().plot(kind ='bar',width = 0.8, stacked= "True")
ax1.legend(loc="upper left", bbox_to_anchor=(1.05,1),fancybox = True, shadow = True, ncol=1)
plt.xlabel('Phylum')
plt.ylabel('% reads')
plt.xticks((0,1,2,3),('Crenarchaeota','Euryarchaeota','Nanoarchaeota', 'Thaumarchaeota'),rotation = 45)
```
